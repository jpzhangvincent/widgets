```python
import numpy as np
```

```python
def compute_graph(amplitude):
    t = np.linspace(0, 4*np.pi, 200);
    y = amplitude*np.sin(t)
    return [list(t), list(y)]
```

```python
%%HTML

Amplitude: <input id="input" type="text">
<div id="output"></div>

<script type="text/Javascript">

    $('#input').change(function(value){
        IPython.notebook.kernel.execute('compute_graph('+parseFloat($('#input').val())+')', 
        {
            iopub: {
                output: function(output_object){
                    jvalue = output_object.content.data['text/plain'];
                    value = JSON.parse(jvalue);
                    x = value[0];
                    y = value[1];
                    // $('#output').text(value);
                    $('#igraph')[0].contentWindow.postMessage(
                                {restyle: {x:[x], y:[y]}},
                                'https://plot.ly');
                }
            }
        }, 
        {
            silent: false
        });        
    })

</script>
```
