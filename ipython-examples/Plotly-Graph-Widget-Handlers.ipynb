## Plotly graph widgets for custom interactivity in IPython notebooks

```python
import json
import time

from __future__ import print_function # For py 2.7 compat

from IPython.html import widgets # Widget definitions
from IPython.display import display, clear_output # Used to display widgets in the notebook
from IPython.utils.traitlets import Unicode # Used to declare attributes of our widget
```

### Some IPython widget source code that will be included in the plotly-python api

```python
%%javascript

window.genUID = function() {
    return 'xxxxxxxx-xxxx-4xxx-yxxx-xxxxxxxxxxxx'.replace(/[xy]/g, function(c) {
        var r = Math.random()*16|0, v = c == 'x' ? r : (r&0x3|0x8);
        return v.toString(16);
    });
};

require(["widgets/js/widget"], function(WidgetManager){

    var GraphView = IPython.DOMWidgetView.extend({
        render: function(){
            var that = this;

            var uid = window.genUID();
            var frameId = uid;
            var loadingId = 'loading-'+uid;

            that.$graph = $('<iframe id="'+frameId+'" src="https://plot.ly/~chris/3375" seamless width="800" height="600"></iframe>');
            that.$graph.appendTo(that.$el);

            that.$loading = $('<div id="'+loadingId+'">Initializing...</div>')
                            .appendTo(that.$el);

            if(!('registeredFrames' in window)){
                window.registeredFrames = [frameId];
            } else {
                var index = window.registeredFrames.indexOf(frameId);
                if (index > -1) {
                    window.registeredFrames.splice(index, 1);
                }
                window.registeredFrames.push(frameId);
            }

            if(!('pingers' in window)){
                window.pingers = {};
            }

            window.pingers[frameId] = setInterval(function(){
                console.log('posting ping');
                that.graphContentWindow = $('#'+frameId)[0].contentWindow;
                that.graphContentWindow.postMessage({ping: true}, 'https://plot.ly');
            }, 2000);


            // Assign a message listener to each widget instantiation
            // and filter the messages by iframe src so that the right message
            // gets passed to the right widget

            if(!('messageListeners' in window)){
                 window.messageListeners = {};
            }

            window.messageListeners[frameId] = function(e) {

                // TODO: Template 'plot.ly' for enterprise
                if(e.origin.indexOf('plot.ly')>-1) {

                    // TODO: Make sure this works for frames with identical src
                    window.registeredFrames.forEach(function(thisFrameId){
                        var frame = document.getElementById(thisFrameId);
                        if(frame === null){
                            window.removeEventListener('message', window.messageListeners[frameId]);
                        }
                        else if(frame.contentWindow === e.source) {
                            console.log(thisFrameId + 'is equal to e.source!');
                            var thisContentWindow = $('#'+thisFrameId)[0].contentWindow; // TODO: rm jquery?
                            var message = e.data;

                            if(message==='pong') {
                                console.log('pong received, ready to receive');
                                $('#loading-'+thisFrameId).hide();
                                clearInterval(window.pingers[thisFrameId]);
                                that.send({event: 'pong', graphId: thisFrameId});
                                thisContentWindow.postMessage({listen: ['zoom', 'click', 'hover']}, 'https://plot.ly');
                            } else if (message.type==='hover' ||
                                       message.type==='zoom'  ||
                                       message.type==='click' ||
                                       message.type==='unhover') {
                                console.log('sending ' + message.type + ' to ' + thisFrameId);
                                that.send({event: message.type, message: message, graphId: thisFrameId});
                            }
                        }
                    });
                }
            };

            // Assign only a single listener from each frame
            // Note: there are multiple listeners to 'message', but messages
            // will are filtered by iframe above.
            window.removeEventListener('message', window.messageListeners[frameId]);
            window.addEventListener('message', window.messageListeners[frameId]);

        },

        update: function() {
            var message = this.model.get('value');

            message = JSON.parse(message);

            var plot = $('#'+message.graphId)[0].contentWindow;
            plot.postMessage(message, 'https://plot.ly');

            return GraphView.__super__.update.apply(this);
        }
    });

    // Register the DatePickerView with the widget manager.
    WidgetManager.register_widget_view('GraphView', GraphView);
});

//@ sourceURL=dynamicScript.js
```

```python
class GraphWidget(widgets.DOMWidget):
    _view_name = Unicode('GraphView', sync=True)
    value = Unicode(sync=True)
    description = Unicode(sync=True)

    def __init__(self, **kwargs):
        super(GraphWidget, self).__init__(**kwargs)

        self._click_handlers = widgets.CallbackDispatcher()
        self._hover_handlers = widgets.CallbackDispatcher()
        self._zoom_handlers = widgets.CallbackDispatcher()
        
        self.on_msg(self._handle_msg)
    
    def restyle(self, data, traces=None):
    
        message = {'restyle': data, 'graphId': self._graphId}
        if traces:
            message['traces'] = traces

        self.value = json.dumps(message)

    def relayout(self, layout):
        message = {'relayout': layout, 'graphId': self._graphId}
        self.value = json.dumps(message)

    def hover(self, hover_obj):
        message = {'hover': hover_obj, 'graphId': self._graphId}
        self.value = json.dumps(message)
        
    def _handle_msg(self, message):
        """Handle a msg from the front-end.
        Parameters
        ----------
        content: dict
            Content of the msg."""
        content = message['content']['data']['content']
        if content.get('event', '') == 'pong':
            self._graphId = content['graphId']
        if content.get('event', '') == 'click':
            self._click_handlers(self, content)
        elif content.get('event', '') == 'hover':
            self._hover_handlers(self, content)
        elif content.get('event', '') == 'zoom':
            self._zoom_handlers(self, content)    
    
    def on_click(self, callback, remove=False):
        """Register a callback to execute when the graph is clicked.
        Parameters
        ----------
        remove : bool (optional)
            Set to true to remove the callback from the list of callbacks."""
        self._click_handlers.register_callback(callback, remove=remove)
    
    def on_hover(self, callback, remove=False):
        """Register a callback to execute when you hover over points in the graph.
        Parameters
        ----------
        remove : bool (optional)
            Set to true to remove the callback from the list of callbacks."""
        self._hover_handlers.register_callback(callback, remove=remove)
        
    def on_zoom(self, callback, remove=False):
        """Register a callback to execute when you zoom in the graph.
        Parameters
        ----------
        remove : bool (optional)
            Set to true to remove the callback from the list of callbacks."""
        self._zoom_handlers.register_callback(callback, remove=remove)

```

### Widget usage

Assign custom callback handlers to events propagated from events fired from Plotly graphs (hover, zoom, and click)

```python
# Custom callback handler

def message_handler(widget, msg):
    clear_output()
    display(msg)
```

```python
g = GraphWidget()           # Custom Plotly "graph widget"
display(g)                  # Display the graph in the notebook
g.on_click(message_handler) # Assign the `message_handler` to the `click` events fired from clicking on points in the graph
g.on_hover(message_handler) # Assign the `message_handler` to the `hover` events fired from hovering over points in the graph
g.on_zoom(message_handler)  # Assign the `message_handler` to the `zoom` events fired from zooming into regions in the graph
```

#### graph widget methods

`restyle` updates the traces in Plotly graphs - update the data or the style of a Plotly graph embedded in an IPython notebook. 

These methods communicate with the Plotly graph locally - no further communication to the Plotly server is required.

```python
# Update the graph's data
g.restyle({'x': [[1,2,3], 
               [1, 2, 4]], 
         'y': [[10,20,30],
               [5, 8, 14]]
}, traces=[0, 1])
```

```python
# Update the graph's style
g.restyle({
    'marker.color': ['rgb(0, 67, 88)', 'rgb(31, 138, 112)']
}, traces=[0, 1])
```

`relayout` updates a graph's `layout`

```python
# Update the graph's layout
g.relayout({'title': 'trials'})
```

`hover` attaches the 'hover' labels to points in a Plotly graph

```python
# Manual hover over points
g.hover({'xval': 2, 'yval': 8})
```

here is a sequence of graph edits, showing the different things about Plotly graphs that you can edit.

```python
# Show a sequence of graph edits

def info_popup(info):
    popup = widgets.PopupWidget(
        description='Scroll up to see the embedded plotly graph update',
        children=[widgets.HTMLWidget(value=json.dumps(info,indent=4).replace('\n', '<br>').replace(' ', '&nbsp'))]
    )
    popup.set_css('display', 'none', selector='.btn')
    return popup

style_sequence = [
    # Clear graph
    {
        'data': {
            'x': [],
            'y': [],
            'marker.color': '',
            'marker.size': '',
            'marker.symbol': '',
            'line.color': '',
            'line.width': '',
            'name': ''
        },
        'layout': {
            'title': '',
            'showlegend': False,
            'legend.x': '',
            'legend.y': '',
            'legend.bgcolor': ''
        },
        'hover': {}
    },
    # Add data
    {
        'data': {
            'x': [[1,2,3], 
                  [1, 2, 4]], 
         'y': [[10,20,30],
               [5, 8, 14]]
        },
        'traces': [0, 1]
    },
    # Restyle markers
    {
        'data': {
            'marker.color': ['rgb(0, 67, 88)', 'rgb(31, 138, 112)'],
            'marker.size': 10,
            'marker.symbol': 'hexagon-open'
        },
        'traces': [0, 1]
    },
    # Restyle lines
    {
        'data': {
            'line.color': ['rgb(0, 67, 88)', 'rgb(31, 138, 112)'],
            'line.width': 3
        }
    },
    # Add title
    {
        'layout': {
            'title': 'Trials',
        }
    },
    # Add legend
    {
        'layout': {
            'showlegend': True,
        }
    },
    # Add legend names
    {
        'data': {
            'name': ['Control', 'Experiment']
        },
        'traces': [0, 1]
    },
    # Move legend
    {
        'layout': {
            'legend.x': 0.02,
            'legend.y': 1,
            'legend.bgcolor': 'rgba(255, 255, 255, 0.5)'
        }
    },
    {'hover': {'xval': 2, 'yval': 8}},
    {'hover': {'xval': 2, 'yval': 8, 'hovermode': 'closest'}},
    {'xval': 2, 'yval': 8, 'hovermode': 'x'},
    {'hover': [{'curveNumber': 0, 'pointNumber': 0}]},
    {'hover': [{'curveNumber': 0, 'pointNumber': 0},
     {'curveNumber': 0, 'pointNumber': 1}]},
    {'hover': [{'curveNumber': 0, 'pointNumber': 2},
     {'curveNumber': 1, 'pointNumber': 1}]},
    {'hover': [{'curveNumber': 0, 'xval': 1, 'yval': 10},
     {'curveNumber': 0, 'xval': 2, 'yval': 20},
     {'curveNumber': 0, 'xval': 3, 'yval': 30},
     {'curveNumber': 1, 'xval': 1, 'yval': 5},
     {'curveNumber': 1, 'xval': 2, 'yval': 8},
     {'curveNumber': 1, 'xval': 4, 'yval': 14}
    ]}
]

for s in style_sequence:
    popup = info_popup(s)
    display(popup)
    if 'data' in s:
        if 'traces' in s:
            g.restyle(s['data'], traces=s['traces'])
        else:
            g.restyle(s['data'])
    if 'layout' in s:
        g.relayout(s['layout'])
    if 'hover' in s:
        g.hover(s['hover'])
    time.sleep(2.5)

    popup.close()
    clear_output()
```

with replotting, you can 'stream' data to plotly graphs locally, without sending data over the network

```python
# Update the graph in real-time
x = []
y = []
for i in range(50):
    x.append(i)
    y.append(i)
    
    g.restyle({'x': [y], 'y': [x]}, traces=[0,1])
    time.sleep(100./1000.)
```
