```python
import pandas as pd
import json
from IPython.display import HTML

import plotly
import plotly.plotly as py
```

```python
df = pd.read_csv('311_150k.csv', infer_datetime_format=True)
df = df
df.head()
```

```python
def restyle(*args, **kwargs):
    
    x_column = args[0]
    y_column = args[1]
    filter_column = args[2]
    
    
    if y_column != "" and filter_column != "":
        idx = df[y_column].str.contains(filter_column).fillna(False)
        dfnew = df[idx]
    else:
        dfnew = df
        
    vc = dfnew[x_column].value_counts()
        
    x = vc.index
    y = vc.values
    text = []
    
    if x_column == 'City' and y_column != "" and filter_column != "":
        complaints_by_city = df['City'].value_counts()
        ynew = []

        for (xi, yi) in zip(x, y):
            ynew.append(100.0*yi/complaints_by_city[xi])
            text.append(yi)
        
        y = ynew
        
        xyt = zip(*sorted(zip(x, y, text), key=lambda z: -z[1]))
        x = xyt[0]
        y = xyt[1]
        text = xyt[2]
        
    restyle = {
        'restyle': {
            'x': [x],
            'y': [y],
            'text': [text],
            'type': 'bar'
        }
    }

    return json.dumps(restyle, cls=plotly.utils._plotlyJSONEncoder)
```

```python
def menu(*items, **options):
    menu_name=options.get('menu_name', '')
    li_class = options.get('li_class', '')

    menu = (''
    '<div class="pure-menu pure-menu-open" style="width: 350px; '
         'display:inline-block; float: left; overflow-y: scroll; max-height: 200px">'
    '    <a class="pure-menu-heading">{menu_name}</a>'
    '    <ul>')
    for item in items:
        menu += '        <li class="{li_class}"><a>{item}</a></li>'.format(li_class=li_class, item=item)

    menu += ('    </ul>'
             '</div>')
    
    menu = menu.format(menu_name=menu_name)
    
    return menu

def search(placeholder, legend, input_class):
    search = (''
    '<form class="pure-form">'
    '    <fieldset>'
    '        <legend>{legend}</legend>'
    '            <input class="{input_class}" type="text" placeholder="{placeholder}">'
    '    </fieldset>'
    '</form>')
    search = search.format(placeholder=placeholder, legend=legend, input_class=input_class)
    return search
```

```python
HTML(menu('index', *df.columns, menu_name='aggregate', li_class='x') +
     menu('index', *df.columns, menu_name='filter', li_class='y') + 
     search('Derelict Vehicles', 'Filter text', 'filter')+
     '<link rel="stylesheet" href="http://yui.yahooapis.com/pure/0.5.0/pure-min.css">')
```

```python
%%HTML

<script type="text/javascript">

var plot = $('#igraph')[0].contentWindow;

var pinger = setInterval(function(){
    console.log('posting ping');
    plot.postMessage({ping: true}, 'https://plot.ly');
}, 5000);

function recompute() {
    var x_column = $('li.x.pure-menu-open').text()
    var y_column = $('li.y.pure-menu-open').text()
    var filter = $('.filter').val()

    restyle(x_column, y_column, filter);    
}

$('.pure-menu li').off('click').on('click', function() {
    if($(this).hasClass('x')) {
        $('li.x').removeClass('pure-menu-open');
    } else if($(this).hasClass('y')) {
        $('li.y').removeClass('pure-menu-open');
    } else if($(this).hasClass('type')) {
        $('li.type').removeClass('pure-menu-open');
    }
    $(this).addClass('pure-menu-open');

    recompute();
});


$('.filter').keyup(recompute);    

function restyle(x, y, type) {
    var exec = 'restyle("{}", "{}", "{}")';
    exec = exec.replace("{}", x);
    exec = exec.replace("{}", y);
    exec = exec.replace("{}", type);
    console.log('exec: ', exec);
    IPython.notebook.kernel.execute(
        exec,
        {
            iopub: {
                output: function(output_object) {
                    if(output_object.msg_type==='pyerr'){
                        console.log('error: ', output_object, output_object.content.evalue);
                    } else {
                        var jvalue = output_object.content.data['text/plain'];
                        var message = JSON.parse(jvalue.substr(1, jvalue.length-2));
                        plot.postMessage(message, 'https://plot.ly');
                    }            
                }
            }
        },
        {
            silent: false
        }
    );    
}
</script>
```
