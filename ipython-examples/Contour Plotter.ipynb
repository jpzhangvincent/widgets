```python
import plotly.plotly as py
from plotly.graph_objs import *
import plotly.tools as tls
import json
import numpy as np
import math
```

```python
def march(x0, y0, x1, y1):
    '''
    Return the closest path of integer coordinates 
    between (x0, y0) and (x1, y1)
    '''
    if abs(x1-x0) > abs(y1-y0):    
        if x1>x0:
            x = range(int(x0), int(x1)+1)
        else:
            x = range(int(x0), int(x1)+1, -1)
        y = []
        tanth = (y1-y0)/(x1-x0)
        for xi in x:
            y.append(round(y0+(xi-x0)*tanth))
    else:
        if y1>y0:
            y = range(int(y0), int(y1)+1)
        else:
            y = range(int(y0), int(y1)+1, -1)
        x = []
        tanth = (x1-x0)/(y1-y0)
        for yi in y:
            x.append(round(x0+(yi-y0)*tanth))

    return (x, y)

class Responder(object):
    '''
    Stateful object that stores and computes the 
    elevation and distance data of the 
    line plot. The 'click' method is executed
    on `click` events of the contour map.
    '''
    def __init__(self):
        self.clicks = 0
        self.x = []
        self.y = []
        self.xp = []
        self.yp = []
        self.z = []
        self.d = []

    def append_point(self, point):
        xp = point['pointNumber'][1]
        yp = point['pointNumber'][0]
        self.xp.append(xp)
        self.yp.append(yp)
        if 'x' in point['data']:
            x = point['data']['x'][xp]
        else:
            x = xp
        if 'y' in point['data']:
            y = point['data']['y'][yp]
        else:
            y = yp

        self.x.append(x)
        self.y.append(y)
        self.z.append(point['z'])
    
        if len(self.x) == 1:
            self.d.append(0)
        else:
            self.d.append(math.sqrt((y-self.y[-2])**2+(x-self.x[-2])**2))
        
        
    def click(self, jdata):
        clickdata = json.loads(jdata);
        point = clickdata['points'][0]
        if self.clicks==0 or self.clicks > 5:
            self.__init__()
            self.append_point(point)
        else:
            
            (xpath, ypath) = march(self.xp[-1], self.yp[-1],
                                   point['pointNumber'][1], point['pointNumber'][0])
    
            d = []
            z = []
            for i in range(1, len(xpath)):
                xpi = xpath[i]
                ypi = ypath[i]

                if 'x' in point['data']:
                    xi = point['data']['x'][xpi]
                else:
                    xi = xpi
                if 'y' in point['data']:
                    yi = point['data']['y'][ypi]
                else:
                    yi = ypi
                
                self.d.append(self.d[-1]+math.sqrt((yi-self.y[-1])**2+(xi-self.x[-1])**2))
                self.z.append(point['data']['z'][int(ypi)][int(xpi)])
                self.x.append(xi)
                self.y.append(yi)
                self.xp.append(xpi)
                self.yp.append(ypi)
            

        self.clicks+=1            
        data = {'restyle': {'x': [self.d], 'y': [self.z]}, 'target': 'plot2'}

        return json.dumps(data)
    
r = Responder()
```

```python
%%HTML   

<h3>Click on points in the contour plot. Their value and distance will be plotted in the line plot below. 
    This notebook requires an IPython backend, so you must run this notebook on your own machine!
</h3>

<!--Contour plot -- replace this with any Plotly contour plot-->
<iframe id="plot1" scrolling="no" style="border:none;" seamless="seamless"
         src="https://plot.ly/~bronsolo/63.embed" height="600" width="100%"></iframe>
<!--Empty line plot-->
<iframe id="plot2" scrolling="no" style="border:none;" seamless="seamless"
         src="https://plot.ly/~chris/2150.embed" height="600" width="100%"></iframe>
    
<script type="text/Javascript">
    window.removeEventListener('message');
    if(typeof(pinger) !== 'undefined') {
        clearInterval(pinger);
    }

    var plot1 = $('#plot1')[0].contentWindow; // Contour plot
    var plot2 = $('#plot2')[0].contentWindow; // line plot
    
    // Initialize communication with Plotly through the iFrame
    var pinger = setInterval(function(){
        console.log('posting ping');
        plot1.postMessage({ping: true}, 'https://plot.ly');
    }, 10000);

    var timeOut;
    window.addEventListener('message', function(e) {
        var message = e.data;
        if(message==='pong') {
            clearInterval(pinger);
            console.log('pong received - assigning listeners');
            // Assign listeners on 'zoom', 'click', and 'hover' events
            plot1.postMessage({listen: ['zoom', 'click', 'hover']}, 'https://plot.ly');
        } else if (message.type==='hover' ||
                   message.type==='zoom' ||
                   message.type==='click' ||
                   message.type==='unhover') {
            if(message.type==='click'){
                // On click events, execute `r.click` in the IPython console
                console.log(message.type, ': ', message);
                var exec_string = 'r.'+message.type+'(\'{}\')'.replace('{}', JSON.stringify(message));
                // console.log('exec: ', exec_string);
                clearTimeout(timeOut);
                timeOut = setTimeout(function(){
                    IPython.notebook.kernel.execute(
                        exec_string,
                        {
                            iopub: {
                                output: process_output
                            }
                        },
                        {
                            silent: false
                        }
                    );
                }, 500);
            }
        }
    });

    var process_output = function(output_object){
        // Callback from IPython.notebook.kernel.execute
        // Make plot commands here
        if(output_object.msg_type==='pyerr'){
            console.log('error: ', output_object, output_object.content.evalue);
        } else {
            var jvalue = output_object.content.data['text/plain'];
            var message = JSON.parse(jvalue.substr(1, jvalue.length-2));
            var targetPlot;
            if(message.target==='plot1') {
                targetPlot = plot1;
            } else if(message.target==='plot2') {
                targetPlot = plot2;
            }
            targetPlot.postMessage(message, 'https://plot.ly');
        }
    };
</script>
```
