```python
import apachelog
import sys
import urllib2
import pandas as pd
import numpy as np
from pandas import Series, DataFrame, Panel
import plotly.plotly as py
```

```python
fformat = r'%V %h %l %u %t \"%r\" %>s %b \"%i\" \"%{User-Agent}i\" %T'
p = apachelog.parser(fformat)
sample_string = 'koldunov.net 85.26.235.202 - - [16/Mar/2013:00:19:43 +0400] "GET /?p=364 HTTP/1.0" 200 65237 "http://koldunov.net/?p=364" "Mozilla/5.0 (Windows NT 5.1) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11" 0'
data = p.parse(sample_string)
log = urllib2.urlopen('https://raw.githubusercontent.com/koldunovn/nk_public_notebooks/master/access_log_for_pandas')
log_list = []
for line in log:
       try:
          data = p.parse(line)
       except:
          sys.stderr.write("Unable to parse %s" % line)
       data['%t'] = data['%t'][1:12]+' '+data['%t'][13:21]+' '+data['%t'][22:27]
      
       log_list.append(data)
```

```python
df = DataFrame(log_list)
del df['%T']; del df['%V']; del df['%i']; del df['%l']; del df['%u']; del df['%{User-Agent}i']
df = df.rename(columns={'%>s': 'Status', '%b':'b', 
                        '%h':'IP', '%r':'Request', '%t': 'Time'})
def dash2nan(x):
    if x == '-':
        x = np.nan
    else:
        x = float(x)/1048576.
    
    return x
df['b'] = df['b'].apply(dash2nan)
df.index = pd.to_datetime(df.pop('Time'))
df.head()
```

```python
def compute(value):
    indx = df['Request'].str.contains(value)
    most_popular_requests = df['Request'][indx].value_counts()[0:30]
    requests = df[indx]['b'].resample('5t', how='count')
    return [list(most_popular_requests.index),
            list(most_popular_requests), 
            list(requests.index.map(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))),
            list(requests)]
```
